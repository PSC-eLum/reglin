```python
def rms(Y1,Y2):
    return np.sqrt(np.mean((Y1 - Y2)**2))
    
def randomforestcross(X,Y):
    X_train, X_1, y_train, y_1 = cross_validation.train_test_split(X, Y, test_size=0.4, random_state=0)
    X_cr, X_test, y_cr, y_test = cross_validation.train_test_split(X_1, y_1, test_size=0.5, random_state=0)
    RMS=np.zeros((100,2))
    
    for i in range(1,25):
        y_=randomforest(i,X_train,y_train,X_cr)
        r=rms(y_,y_cr)
        RMS[i,0]=i
        RMS[i,1]=r
    plt.scatter(RMS[:,0],RMS[:,1])
    plt.show()
    L = [X_train, y_train, X_test, y_test]
    return L 
def randomforest(n,X,Y,X_t):
    clf= RandomForestRegressor(n_estimators=n)
    clf.fit(X,Y)
    y_=clf.predict(X_t)
    return y_

def gradientcross(X,Y):
    X_train, X_1, y_train, y_1 = cross_validation.train_test_split(X, Y, test_size=0.4, random_state=0)
    X_cr, X_test, y_cr, y_test = cross_validation.train_test_split(X_1, y_1, test_size=0.5, random_state=0)
    RMS=np.zeros((250,2))
    
    for i in range(1,10):
        i = i * 10 + 100
        y_=gradient(i,X_train,y_train,X_cr)
        r=rms(y_,y_cr)
        RMS[i,0]=i
        RMS[i,1]=r
    plt.scatter(RMS[:,0],RMS[:,1])
    plt.show()
    L = [X_train, y_train, X_test, y_test]
    return L 
def gradient(n,X,Y,X_t):
    clf= ensemble.GradientBoostingRegressor(n_estimators=n)
    clf.fit(X,Y)
    y_=clf.predict(X_t)
    return y_

    

  
```
